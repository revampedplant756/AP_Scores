```python
def District_Cleaner(df, check):
  df = df.drop(columns=["REGION", "COUNTY NAME", "DISTRICT NAME", "CHARTER STATUS", "YEAR"], axis=1)
  df = df.loc[df["GRADE"] > 8]
  df = df.drop(columns=["GRADE"], axis=1)
  if check > 5:
    df["ENROLLMENT"] = df["ENROLLMENT"].str.replace(',', '').astype(float)
  df["ENROLLMENT"] = pd.to_numeric(df["ENROLLMENT"])
  df = df.groupby(['DISTRICT'], as_index=False).sum()



  return df
```
